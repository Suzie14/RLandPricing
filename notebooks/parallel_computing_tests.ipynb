```python
from joblib import Parallel, delayed

def process(index, beta, nb_games):
    game = interact.Interaction(beta=beta)
    result = game(nb_iterations=nb_games)
    result['beta'] = beta
    result['index'] = index
    return result

betas = sorted([10**(-3), 10**(-5)]*2)
    
results = Parallel(n_jobs=4)(delayed(process)(index, beta, 10**5) for index, beta in enumerate(betas))
results_df = pd.concat(results, ignore_index=True)
print(results_df)  
```
